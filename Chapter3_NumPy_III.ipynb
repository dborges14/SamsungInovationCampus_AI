Again, use the dot() method or function of NumPy to do the matrix multiplication.
The number of columns of the left matrix = the number of rows of the right matrix. 

```python
m1 = np.array([[1, 2, 3],[4, 5, 6]])	# A 2 x 3 matrix.			
m2 = np.array([[6, 5, 4],[3, 2, 1]])	# Another 2 x 3 matrix.
print('M1 = ', m1,'\n')
print('M2 = ', m2,'\n')

np.transpose(m2)  		# Transpose a 2 x 3 matrix to get another of shape 3 x 2.
```

```python
np.dot(m1 , np.transpose(m2))		# Multiply 2 x 3 matrix by another of shape 3 x 2.
                                # The result is a matrix of shape 2 x 2.
```

```python
A = np.array([[5, 8],[6, 4]])			# The matrix 'A' which is of square shape.
b = np.array([[30], [25]])			
print(A)
print(b)
```

In linear algebra, an *n-by-n* square matrix  is called invertible, if there exists an *n-by-n* square matrix $A^{-1}$ such that

#$AA^{-1}=I_n$

where $I_n$ denotes the *n-by-n* identity matrix and the multiplication used is ordinary matrix multiplication.

For *2-by-2* matrices, a simple way of calculating the inverse is as follows

image.png

```python
np.linalg.det(A)
```

```python
Ainv = np.linalg.inv(A)			# The inverse of 'A'.
Ainv
```

```python
np.dot(Ainv,b)		 		# The solution by matrix multiplication.
```

```python
np.linalg.solve(A, b)	 		# The solution by using np.linalg.solve() function.
```
